\section*{Вычисление производной $y = x^T x$ по вектору $x$}


Дана скалярная функция от вектора:
$$
y = x^T x, \quad x \in \mathbb{R}^N,
$$
где $x$ — вектор‑столбец размерности $N \times 1$.


\subsection*{Шаг 1. Раскрытие выражения для $y$}


Запишем $x$ в координатной форме:
$$
x = \begin{pmatrix} x_1 \\ x_2 \\ \vdots \\ x_N \end{pmatrix}.
$$

Тогда:
$$
y = x^T x = \sum_{i=1}^N x_i^2 = x_1^2 + x_2^2 + \cdots + x_N^2.
$$

\subsection*{Шаг 2. Вычисление частных производных}

Градиент $\frac{dy}{dx}$ — это вектор частных производных:
$$
\frac{dy}{dx} = \begin{pmatrix}
\frac{\partial y}{\partial x_1} \\
\frac{\partial y}{\partial x_2} \\
\vdots \\
\frac{\partial y}{\partial x_N}
\end{pmatrix}.
$$

Найдём $\frac{\partial y}{\partial x_j}$:
$$
\frac{\partial y}{\partial x_j} = \frac{\partial}{\partial x_j} \left( \sum_{i=1}^N x_i^2 \right) = \frac{\partial}{\partial x_j} (x_j^2) = 2x_j.
$$
(Остальные слагаемые не зависят от $x_j$, их производные равны нулю.)

\subsection*{Шаг 3. Сборка градиента}

Подставляем частные производные в вектор:
$$
\frac{dy}{dx} = \begin{pmatrix}
2x_1 \\
2x_2 \\
\vdots \\
2x_N
\end{pmatrix} = 2 \begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_N
\end{pmatrix} = 2x.
$$

\subsection*{Ответ}

Таким образом, производная равна:
$$
\boxed{\frac{dy}{dx} = 2x}.
$$

\subsection*{Альтернативный вывод (через матричное дифференцирование)}

Для квадратичной формы $y = x^T A x$ известно:
$$
\frac{dy}{dx} = (A + A^T)x.
$$

В нашем случае $A = I$ (единичная матрица), поэтому:
$$
\frac{dy}{dx} = (I + I^T)x = 2Ix = 2x.
$$

\section*{Вычисление производной $y = \operatorname{tr}(AB)$ по матрице $A$}

Дана скалярная функция от матриц:
$$
y = \operatorname{tr}(AB), \quad A, B \in \mathbb{R}^{N \times N},
$$
где $\operatorname{tr}(\cdot)$ — операция взятия следа матрицы (суммы диагональных элементов).

\subsection*{Шаг 1. Раскрытие определения следа}

След матрицы $AB$ можно записать как:
$$
y = \operatorname{tr}(AB) = \sum_{i=1}^N (AB)_{ii},
$$
где $(AB)_{ii}$ — элемент на позиции $(i,i)$ в произведении матриц $AB$.

\subsection*{Шаг 2. Выражение элементов произведения матриц}

Элемент $(AB)_{ii}$ вычисляется как:
$$
(AB)_{ii} = \sum_{k=1}^N A_{ik} B_{ki}.
$$

Тогда:
$$
y = \sum_{i=1}^N \sum_{k=1}^N A_{ik} B_{ki}.
$$

\subsection*{Шаг 3. Вычисление частной производной}

Найдём частную производную $y$ по элементу $A_{pq}$:
$$
\frac{\partial y}{\partial A_{pq}} = \frac{\partial}{\partial A_{pq}} \left( \sum_{i=1}^N \sum_{k=1}^N A_{ik} B_{ki} \right).
$$

Заметим, что в двойной сумме только одно слагаемое зависит от $A_{pq}$ — это слагаемое при $i = p$ и $k = q$:
$$
\frac{\partial y}{\partial A_{pq}} = \frac{\partial}{\partial A_{pq}} (A_{pq} B_{qp}) = B_{qp}.
$$

\subsection*{Шаг 4. Сборка матрицы производных}

Матрица производных $\frac{dy}{dA}$ имеет элементы:
$$
\left( \frac{dy}{dA} \right)_{pq} = \frac{\partial y}{\partial A_{pq}} = B_{qp}.
$$

Это означает, что:
$$
\frac{dy}{dA} = B^T,
$$
так как элемент $(p,q)$ в $\frac{dy}{dA}$ равен элементу $(q,p)$ в $B$.

\subsection*{Ответ}

Таким образом, производная равна:
$$
\boxed{\frac{dy}{dA} = B^T}.
$$

\subsection*{Альтернативное обоснование (через свойства следа)}

Используем известное свойство следа:
$$
\operatorname{tr}(AB) = \operatorname{tr}(BA),
$$
и правило дифференцирования следа произведения:
$$
d\,\operatorname{tr}(AB) = \operatorname{tr}\left( (dA) B \right).
$$

Сравнивая с общим видом дифференциала:
$$
dy = \operatorname{tr}\left( \left(\frac{dy}{dA}\right)^T dA \right),
$$
получаем:
$$
\left(\frac{dy}{dA}\right)^T = B \quad \Rightarrow \quad \frac{dy}{dA} = B^T.
$$

\subsubsection{Шаг 1. Раскрытие скалярного произведения}

Запишем $y$ в координатной форме:
$$
y = x^T (A c) = \sum_{i=1}^N x_i \cdot (A c)_i,
$$
где $(A c)_i$ — $i$-й элемент вектора $A c$.

\subsubsection{Шаг 2. Вычисление частной производной}

Найдём частную производную по $x_j$:
$$
\frac{\partial y}{\partial x_j} = \frac{\partial}{\partial x_j} \left( \sum_{i=1}^N x_i (A c)_i \right) = (A c)_j,
$$
так как только слагаемое с $i = j$ зависит от $x_j$.

\subsubsection{Шаг 3. Сборка градиента}

Градиент — это вектор частных производных:
$$
\frac{dy}{dx} = \begin{pmatrix}
\frac{\partial y}{\partial x_1} \\
\frac{\partial y}{\partial x_2} \\
\vdots \\
\frac{\partial y}{\partial x_N}
\end{pmatrix} = \begin{pmatrix}
(A c)_1 \\
(A c)_2 \\
\vdots \\
(A c)_N
\end{pmatrix} = A c.
$$

\subsubsection{Ответ для $\frac{dy}{dx}$}
$$
\boxed{\frac{dy}{dx} = A c}.
$$

\subsection*{2. Производная по матрице $A$: $\frac{dy}{dA}$}

\subsubsection{Шаг 1. Представление через след}

Заметим, что:
$$
y = x^T A c = \operatorname{tr}(x^T A c).
$$
Используя свойство следа $\operatorname{tr}(ABC) = \operatorname{tr}(CAB)$, перепишем:
$$
y = \operatorname{tr}(c x^T A).
$$
Здесь $c x^T$ — матрица размера $N \times N$ (внешнее произведение векторов).

\frac{dy}{dA} = M^T.

$$
\frac{dy}{dA} = (c x^T)^T = x c^T.

- $x \in \mathbb{R}^N$ (вектор‑столбец),
- $c^T \in \mathbb{R}^{1 \times N}$ (вектор‑строка),
- их произведение $x c^T$ — матрица $N \times N$, что соответствует размерности $A$.

\subsubsection{Ответ для $\frac{dy}{dA}$}
$$
\boxed{\frac{dy}{dA} = x c^T}.
$$

## Question 3

Which of the following statements about $k$-Nearest Neighbor ($k$-NN) are true in a classification setting, and for all $k$? Select all that apply.
1. The decision boundary (hyperplane between classes in feature space) of the k-NN classifier is linear.
2. The training error of a 1-NN will always be lower than that of 5-NN.
3. The test error of a 1-NN will always be lower than that of a 5-NN.
4. The time needed to classify a test example with the k-NN classifier grows with the size of the training set.
5. None of the above.

*Your Answer:*


*Your Explanation:*




«The decision boundary (hyperplane between classes in feature space) of the k‑NN classifier is linear»
**Неверно**.

Граница решений у k‑NN нелинейна и адаптивна: она зависит от расположения обучающих примеров и значения k.

Для k = 1 граница «огибает» точки, образуя сложные, часто извилистые участки.

Линейная граница (гиперплоскость) характерна, например, для логистической регрессии или SVM с линейным ядром, но не для k‑NN.

«The training error of a 1‑NN will always be lower than that of 5‑NN»
**Верно**.

При k = 1 классификатор всегда присваивает тестовому примеру метку самого близкого соседа. На обучающей выборке это означает: каждый пример «сравнивается» сам с собой (расстояние = 0), поэтому ошибка на обучении равна 0 (если нет дубликатов с разными метками).

При k = 5 голосование ведётся по 5 соседям; возможны случаи, когда среди них есть примеры других классов — это может увеличить ошибку на обучающей выборке.

Таким образом, ошибка обучения у 1‑NN гарантированно ≤ ошибки 5‑NN (чаще строго меньше).

«The test error of a 1‑NN will always be lower than that of a 5‑NN»
**Неверно**.

На тестовых данных 1‑NN склонен к переобучению: он слишком чувствителен к шуму и выбросам, так как учитывает только одного ближайшего соседа.

5‑NN «сглаживает» предсказания за счёт голосования по 5 соседям, что часто снижает обобщающую способность и тестовую ошибку.

В реальности 5‑NN нередко даёт меньшую ошибку на тесте, чем 1‑NN. Утверждение «всегда ниже» ложно.

«The time needed to classify a test example with the k‑NN classifier grows with the size of the training set»
**Верно**.

k‑NN — метод «ленивого обучения»: на этапе предсказания он вычисляет расстояния от тестового примера до всех точек обучающей выборки.

Чем больше обучающий набор, тем больше расстояний нужно посчитать → время классификации растёт (обычно линейно от размера обучающей выборки).

Оптимизации (KD‑деревья, Ball‑деревья) могут ускорить поиск соседей, но в общем случае зависимость от размера выборки сохраняется.
